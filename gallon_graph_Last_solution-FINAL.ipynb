```python
class G(object):
    def __init__(self, volm,vola):
        if vola>volm:
            raise Exception('out of volume')
        self.vola = vola
        self.volm = volm
        if self.vola<self.volm:
            self.etat = False
        else:
            self.etat =True
        if self.vola==0:
            self.vide = True
        else:
            self.vide = False
    def remp(self,other):
        if not other.etat and not self.vide:
            aco = other.vola
            acs = self.vola
            other.vola=(aco+acs)*(aco+acs<other.volm)+(other.volm)*(aco+acs>=other.volm)
            self.vola=((acs+aco)-other.vola)*((acs+aco)-other.vola>0)                
    def __add__(self,other):
        if not other.etat and not self.vide:
            aco = other.vola
            acs = self.vola
            other.vola=(aco+acs)*(aco+acs<other.volm)+(other.volm)*(aco+acs>=other.volm)
            self.vola=((acs+aco)-other.vola)*((acs+aco)-other.vola>0)
```

```python
def sol(target,ini):
    global path
    sols=set()
    new_nodes=niv(*ini,ini)[0]
    while new_nodes and not sols:
        temp=set()
        for node in new_nodes:
            temp.update(niv(*node,gv(node,path))[0])
        new_nodes=temp
        for u in new_nodes:
            if target in u :
                sols.update({gv(u,path)})
    return sols
```

```python
def tree(ini):
    global path 
    new_nodes=niv(*ini,ini)[0]
    while new_nodes:
        temp=set()
        for node in new_nodes:
            temp.update(niv(*node,gv(node,path))[0])
        new_nodes=temp
    return path
```
